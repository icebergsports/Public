```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

def print_teams(): 

    print("Team names for Input: ")
    print("Atlanta Braves - ATL ")
    print("Miami Marlins - MIA ") 
    print("New York Mets - NYM ")
    print("Philadelphia Phillies - PHI")
    
    print("Washington Nationals - WSH")
    print("Chicago Cubs - CHC")
    print("Cincinnati Reds - CIN")
    print("Milwaukee Brewers -MIL")
    print("Pittsburgh Pirates - PIT")
    print("St. Louis Cardinals - STL")

    print("Arizona Diamondbacks - ARI")
    print("Colorado Rockies - COL")
    print("Los Angeles Dodgers - LAD")
    print("San Diego Padres - SD")
    print("San Francisco Giants - SF")

    print("Baltimore Orioles - BAL")
    print("Boston Red Sox - BOS")
    print("New York Yankees  - NYY")
    print("Tampa Bay Rays - TB")
    print("Toronto Blue Jays - TOR")
    print("Chicago White Sox - CHW")
    print("Cleveland Guardians - CLE")
    print("Detroit Tigers - DET")
    print("Kansas City Royals - KC")
    print("Minnesota Twins - MIN")

    print("Houston Astros - HOU")
    print("Los Angeles Angels - LAA")
    print("Oakland Athletics - OAK")
    print("Seattle Mariners -SEA")
    print("Texas Rangers - TEX")
    


def main(): 
    DF = pd.read_csv("/Users/home/Desktop/inningScore.csv") 
    
    team = DF['Team']
    one = DF['1']
    two = DF['2']
    three = DF['3']
    four = DF['4']
    five = DF['5']
    six = DF['6']
    seven = DF['7']
    eight = DF['8']
    nine = DF['9']
    total = DF['R']

    print_teams()

    bankroll = []
    balance = 500
    wins = 0 
    loss = 0


    for i in range(0, len(team)): 
        profit = 0
        betsize = 50

        if i == 0 or i%4 == 0:

            fifth = one[i] + two[i] + three[i] + four[i] + five[i]
            fifth2 = one[i+1] + two[i+1] + three[i+1] + four[i+1] + five[i+1]
            fifth3 = one[i+2] + two[i+2] + three[i+2] + four[i+2] + five[i+2]
            fifth4 = one[i+3] + two[i+3] + three[i+3] + four[i+3] + five[i+3]




            if fifth > 0 and fifth2 > 0 and fifth3 >0 and fifth4 > 0: 

                profit = betsize * 2

            else: 
                profit = betsize * -1


        if profit > 0: 
            wins = wins + 1 

        if profit < 0: 
            loss = loss + 1 

        
        balance = balance + profit 

        bankroll.append(balance)


    plt.plot(bankroll)
    plt.xlabel('Games') 
    plt.ylabel('Bankroll')
    plt.title('MLB Betting Strategy')
    plt.show()

    print("Accuracy: ", wins/(wins+loss)*100,"%")
        
            

    

main() 
```
